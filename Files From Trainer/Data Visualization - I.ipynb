### Adjusting the Line
- You can adjust the line style using the linestyle keyword

```python
# Method
plt.plot(x,np.sin(x),color='green',linestyle='solid')                    #Solid Line
```

```python
# Method 2 
plt.plot(x,np.sin(x),color='yellow',linestyle='dashed')       # Dashed Line
```

```python
# Method 3
plt.plot(x,np.sin(x),color='black',linestyle='dashdot')             #DashDot
```

```python
# Method 4
plt.plot(x,np.sin(x),color='red',linestyle='dotted')             # Dotted
```

### Labelling a Plot
- Labeling of the plot is done : **Grid, Title, Axis, Labels and Legends**

#### Plot Grid
- We can add a grid to the plot by calling the grid() function; it takes one parameter, a Boolean value, to enable (if True) or disable (if False) the grid

```python
# Example 1
t=np.arange(0,5,0.2)
plt.plot(t,t,'r--',t,t**2,'bs')
plt.plot(t,t**3,'g^')
plt.grid(True)
```

### Plot Title and Labels

```python
# Example 1
plt.figure(figsize=(10,5))
x=np.arange(1,5)
y=np.arange(1,5)
plt.plot(x,y,color='magenta',linestyle=':',linewidth=3)
plt.xlabel('X-Axis',color='red',size=20)
plt.ylabel('Y-Axis',color='b',size=20)
plt.title('X-Y Graph',color='k',size=30)
plt.grid(True)
```

### Bar Graph
- A bar chart or bar graph is a chart or graph that presents categorical data with rectangular bars with heights or lengths proportional to the values that they represent. The bars can be plotted vertically or horizontally.

```python
# Example 1
import matplotlib.pyplot as plt
lang=['C','C++','Java','Python','Php']
students=[23,17,35,29,12]
x=plt.bar(lang,students,width=0.5,color='red')

plt.show
```

```python
# Example 2
import numpy as np
N = 5
men_means = (20, 35, 30, 35, 27)
women_means = (25, 32, 34, 20, 25)
ind=np.arange(5)
width=0.3
plt.bar(ind,men_means,width,label='Men',color='black')
plt.bar(ind+width,women_means,width,label='Women',color='blue')
plt.grid(True)
plt.xlabel('Groups')
plt.ylabel('Scores')
plt.title("Scores by Men and Women")

plt.xticks(ind,('G1','G2','G3','G4','G5'),rotation=90)
plt.legend()
plt.show()

# ticks are used to show specific points on the coordinate axis
```

## Stacked Bar Graph

```python
# Example 3
countries=['USA','GB','India','France','Japan']
bronze=np.array([38,17,26,19,15])
silver=np.array([37,23,18,18,11])
gold=np.array([46,27,28,17,19])

ind=[x for x, _ in enumerate(countries)]
plt.bar(ind,gold,width=0.7,label='Gold',color='gold',bottom=silver+bronze)
plt.bar(ind,silver,width=0.7,label="Silver",color='silver',bottom=bronze)
plt.bar(ind,bronze,width=0.7,label='Bronze',color='brown')

plt.xticks(ind,countries)
plt.ylabel=('No of Medal')
plt.xlabel=('Countries')
plt.legend(loc='upper center')
plt.title('Olympic Medal Tally')
plt.grid(False)
```
