```python
import numpy, matplotlib.pyplot as plt
x = numpy.arange(5)
plt.plot(x, x, label='linear')
plt.plot(x, x*x, label='square')
plt.plot(x, x*x*x, label='cube')
plt.grid(True)
plt.xlabel('X-axis')
plt.ylabel('Y-axis')
plt.title('Polynomial Graph')
plt.legend()
plt.show()
```

```python
import seaborn as sns
import matplotlib.pyplot as plt

# Load dataset
iris = sns.load_dataset('iris')

# Use histplot instead of distplot
sns.histplot(iris['petal_length'], 
             label='Petal Length', 
             bins=20,   # number of bins
             kde=False)

plt.legend()
plt.show()
```

```python
import numpy as np
import seaborn as sns
import matplotlib.pyplot as plt

# Load dataset
iris = sns.load_dataset('iris')

# Barplot with NumPy mean as estimator
sns.barplot(
    x='species',
    y='sepal_width',
    data=iris,
    estimator=np.mean,   # <-- correct usage
    palette="spring"
)

plt.show()
```

```python
import seaborn as sns
import matplotlib.pyplot as plt

# Load dataset
flights = sns.load_dataset("flights")

# Pivot correctly
flights = flights.pivot(index="month", columns="year", values="passengers")

# Heatmap
sns.heatmap(
    flights,
    cmap="YlGnBu",
    linewidths=2,
    xticklabels=True,
    yticklabels=False
)

plt.show()
```
