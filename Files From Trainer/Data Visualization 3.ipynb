## Data Visualization using Seaborn

- Seaborn is an open source, licensed python library providing high level API for visualiazing the data using Python Programming Language
- Seaborn helps resolving two major problems faced by Matplotlib
- Default Matplotlib Parameters
- Working with DataFrames

### Important Features of Seaborn

- Visualizing Univariate and Bivariate data
- Visualizing Linear regression models
- Plotting statistical Time-Series Data
- Seaborn works well with Numpy & Pandas Data Structures
- It has built in themes for styling matplotlib graphics

# Importing Libraries

```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
%matplotlib inline
```

```python
x=sns.get_dataset_names()
```

```python
print(x,end=" ")
```

```python
titan=sns.load_dataset('titanic')
titan
```

```python
titan.head(5)
```

```python
sns.countplot(x = 'survived',data = titan)
plt.title('No of Survived Passengers')
plt.xlabel("Survival")
plt.ylabel('Count')
plt.show()
```

```python
survive=sns.countplot(x='pclass',hue='survived',palette='Set3',data=titan)
survive.set(title='Passenger Survival/Death based on Passenger Class',
            xlabel='Passenger Class',ylabel='Count')
plt.legend(loc='upper right')
plt.show()
```

```python
# Loading Tips Dataset
tips=sns.load_dataset('tips')
tips
```

```python
# Bar Graph
sns.barplot(x='day',y='tip',data=tips)
```

```python
# bar graph with hue and palette

sns.barplot(x='day',y='tip',data=tips, hue='sex',palette='Set2')
plt.plot()
```

```python
sns.barplot(x='day',y='total_bill',data=tips, hue='smoker',palette='Blues')
plt.plot()
```

```python
# plot bar graph

sns.barplot(x='day',y='total_bill',data=tips,palette='OrRd',order=['Sat','Sun','Fri','Thur'])
```

```python
# bar plot with median in place of mean

from numpy import median
sns.barplot(x='day',y='total_bill',data=tips,estimator=median)
```

```python
# bar plot with median in place of mean with hue & palette

sns.barplot(x='smoker',y='tip',data=tips,estimator=median,hue='sex',palette='coolwarm')
```

```python
# bar plot with attribute CI

sns.barplot(x='smoker',y='tip',ci=30,data=tips,hue='sex')
```

```python
sns.barplot?
```

```python
# bar plot with capsize

sns.barplot(x='day',y='total_bill',data=tips,palette='spring',capsize=0.1)
```

```python
# barplot with the help of size & tip

sns.barplot(x='size',y='tip',data=tips,palette='Spectral',capsize=0)
```

```python
# Load Iris Dataset

iris=sns.load_dataset('iris')
iris
```

```python
# box plot
plt.figure(figsize=(10,7))
sns.boxplot(data=iris,palette='coolwarm')
plt.xticks(rotation=45)
plt.xlabel('Iris Length')
plt.ylabel('Actual Length')
plt.title('Length of Sepal and Petal')
```
